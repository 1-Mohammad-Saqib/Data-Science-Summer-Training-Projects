```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("image.jpg").convert("RGB")
img = np.array(img)

h, w, _ = img.shape
mid_h = h // 2
mid_w = w // 2

output = img.copy()

output[:mid_h, :mid_w, 0] = np.clip(output[:mid_h, :mid_w, 0] + 80, 0, 255)
output[:mid_h, mid_w:, 1] = np.clip(output[:mid_h, mid_w:, 1] + 80, 0, 255)
output[mid_h:, :mid_w, 2] = np.clip(output[mid_h:, :mid_w, 2] + 80, 0, 255)

gray = np.mean(output[mid_h:, mid_w:], axis=2).astype(np.uint8)
output[mid_h:, mid_w:] = np.stack([gray, gray, gray], axis=2)

plt.figure(figsize=(8, 8))
plt.imshow(output)
plt.axis("off")
plt.title("Image Divided into 4 Parts with Different Effects")
plt.show()
```
