```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
x1 = [2,2,8,8,2]
y1 = [2,6,6,2,2]
x2 = [4,4,4]
y2 = [6,2,6]
x3 = [2,4,3,2,8]
y3 = [6,6,8,6,6]
x4 = [8,7,3]
y4 = [6,8,8]
x5 = [5.5,5.5,6.5,6.5,5.5,6.5]
y5 = [3.5,4.5,4.5,3.5,3.5,4.5]
x6 = [2.5,2.5,3.5,3.5]
y6 = [2,4.5,4.5,2]
circle = Circle([3,6.8],0.3)
```
